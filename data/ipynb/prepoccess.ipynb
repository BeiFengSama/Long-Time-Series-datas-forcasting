一、北京空气质量预处理

```python
import pandas as pd

passenger = pd.read_csv('..\\beijing-airquility\\station-data\\PRSA_Data_Tiantan_20130301-20170228.csv')
# 合并时间列
passenger['Date'] = passenger['year'].astype(str) + '-' + passenger['month'].astype(str) + '-' + passenger['day'].astype(str)

# 将合并后的列转换为日期时间类型
passenger['Date'] = pd.to_datetime(passenger['Date'])

passenger.sort_values(by=['Date'], inplace=True, ascending=True)
```

```python
# 查看数据插补前

from matplotlib.pyplot import figure 
import matplotlib.pyplot as plt 

# 筛选出日期在2000年至2010年之间的数据
filtered_passenger = passenger[(passenger['Date'].dt.year == 2013) & (passenger['Date'].dt.month == 3)]

figure(figsize=(12, 5), dpi=80, linewidth=10) 
plt.plot(filtered_passenger['Date'], filtered_passenger['PM2.5']) 
plt.title('Air Passengers Raw Data with Missing Values') 
plt.xlabel('Years', fontsize=14) 
plt.ylabel('Number of Passengers', fontsize=14) 
plt.show()
```
